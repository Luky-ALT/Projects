# OUTPUTTING AS AN EXCEL **FILE**

```python
import pandas as pd


df1 = pd.DataFrame ({ "employee" : ["Julia", "Marie", "Adam", "Eva", "Nicole"],
        "department" : ["Data Science", "Cyber Security", "Web Development", "Data Science", "Cyber Security"],
        "Year" : ["2005", "2010", "2012", "2010", "2008"]})

df2 = pd.DataFrame ({ "employee" : ["Nicole", "Julia", "Eva", "Adam", "Marie"],
        "Country" : ["Canada", "Canada", "Canada", "USA", "USA"],
        "Salary" : ["2000", "5000", "7000", "2000", "2000"]})
```

```python
df3 = pd.merge(df1, df2)
```

```python
df3
```

```python
df3.to_excel("new_data_set.xlsx", sheet_name = "sheet_name", index = False)
```
